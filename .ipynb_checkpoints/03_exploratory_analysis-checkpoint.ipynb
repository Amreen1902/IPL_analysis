# IPL Exploratory Data Analysis (EDA)

## Objective

The objective of this notebook is to explore IPL historical data and identify patterns related to team performance, player achievements, venue trends, toss outcomes, and match characteristics.

This analysis serves as the foundation for advanced team, player, and predictive analytics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12,6)
```

```python
matches = pd.read_csv('data/cleaned_matches.csv')

deliveries = pd.read_csv('data/cleaned_deliveries.csv')
```

```python
deliveries.head()
```

```python
deliveries.info()
```

## Most Successful IPL Teams

This analysis identifies franchises with the highest number of victories.

```python
team_wins = matches['winner'].value_counts()

team_wins.head(10)
```

```python
plt.bar(team_wins.index, team_wins.values)

plt.xticks(rotation=90)

plt.title("Most Successful IPL Teams")

plt.xlabel("Team")

plt.ylabel("Wins")

plt.show()
```

### Insights

- Mumbai Indians emerged as the most successful franchise in IPL history with **144 victories**, followed closely by Chennai Super Kings with **138 wins**.
- Kolkata Knight Riders secured **131 victories**, establishing themselves among the league's most consistently successful teams.
- The top three franchises collectively recorded **413 wins**, highlighting sustained dominance over multiple IPL seasons.
- Newer franchises such as Gujarat Titans have lower overall win counts because of fewer seasons played rather than poor performance.
- Historical win totals indicate long-term competitive strength and organizational stability across multiple tournament cycles.

### Business Interpretation

Total victories serve as a key performance indicator for franchise success and provide a benchmark for comparing long-term team performance.

## Matches Played Per Season

Analyzing IPL expansion and season growth over time.

```python
matches_per_season = (
    matches.groupby('season_year')
    .size()
)
```

```python
plt.plot(
    matches_per_season.index,
    matches_per_season.values,
    marker='o'
)
plt.title("IPL Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.grid()
plt.show()
```



### Insights

- The number of matches played per season generally increased over time as the IPL expanded and introduced additional franchises.
- Growth in fixture volume reflects the commercial and competitive evolution of the league.
- More recent seasons contribute a larger share of the overall dataset, improving the reliability of modern trend analysis.
- Tournament structure changes have periodically influenced seasonal match counts.

## Toss Winners Distribution

Understanding toss outcomes across IPL history.

```python
toss_wins = matches['toss_winner'].value_counts()
```

```python
plt.bar(
    toss_wins.index,
    toss_wins.values
)

plt.xticks(rotation=90)
plt.title("Toss Wins by Team")
plt.show()
```

### Insights

- Toss wins appear relatively evenly distributed among teams, suggesting no systematic bias.
- Teams with longer IPL participation naturally accumulate more toss wins due to playing more matches.
- Winning the toss alone does not guarantee match success and should be analyzed alongside match outcomes.
- Further analysis is required to determine whether batting first or chasing provides a significant advantage.

### Business Interpretation

Toss outcome is an important strategic factor, but its true impact can only be evaluated when linked to match results.

## Most Frequently Used Venues

```python
top_venues = (
    matches['venue']
    .value_counts()
    .head(10)
)
```

```python
plt.barh(
    top_venues.index,
    top_venues.values
)

plt.title("Top IPL Venues")

plt.xlabel("Matches Hosted")

plt.show()
```

### Insights

- A limited number of venues host a large proportion of IPL matches.
- These venues play a major role in shaping historical IPL statistics.
- Frequent hosting results in larger sample sizes, making venue-specific analysis more reliable.
- Certain venues may exhibit distinct batting-friendly or bowling-friendly characteristics.

### Business Interpretation

Venue analysis can help identify location-specific advantages and support strategic decision-making for teams.

## Top Player of the Match Winners

```python
pom = (
    matches['player_of_match']
    .value_counts()
    .head(10)
)
```

```python
plt.bar(
    pom.index,
    pom.values
)

plt.xticks(rotation=90)

plt.title(
    "Top Player of the Match Winners"
)

plt.show()
```

### Insights

- Ab De Villiers leads the Player of the Match rankings, demonstrating exceptional match-winning impact.
- The top award winners consistently influenced results through batting, bowling, or all-round performances.
- Award concentration among a limited group of players suggests that elite performers often determine match outcomes.
- Player of the Match frequency can serve as a useful proxy for measuring player impact beyond traditional statistics.

## Winning Margin Distribution

```python
plt.hist(
    matches['result_margin'],
    bins=20
)

plt.title(
    "Winning Margin Distribution"
)

plt.xlabel("Margin")

plt.ylabel("Frequency")

plt.show()
```

### Insights

- Most IPL matches are decided by moderate margins, indicating competitive contests.
- Extremely dominant victories occur less frequently.
- A concentration around smaller margins suggests a high level of parity among teams.
- Outlier matches with large victory margins may indicate exceptional individual or team performances.

### Business Interpretation

Winning margin analysis helps measure competitiveness and identify trends in match dominance.

## Runs Scored Per Season

```python
season_runs = (
    deliveries.groupby('match_id')
    ['total_runs_ball']
    .sum()
)

season_runs.describe()
```

### Insights

- The average IPL match produces approximately **318 runs**, indicating the league's high-scoring nature.
- Half of all matches fall between **283 and 354 total runs**, suggesting a relatively consistent scoring environment.
- The highest recorded match total reached **549 runs**, demonstrating the extreme scoring potential of T20 cricket.
- The lowest match total was **56 runs**, highlighting the occasional occurrence of batting collapses and bowling-dominated contests.
- The standard deviation of approximately **59 runs** indicates moderate variation in match scoring patterns.

### Business Interpretation

Run distributions help quantify match competitiveness and provide valuable context for evaluating batting and bowling performance.

# Conclusion

The exploratory analysis of 1,095 IPL matches and 260,920 ball-by-ball records reveals several important patterns within the tournament.

Mumbai Indians emerged as the most successful franchise with 144 victories, followed closely by Chennai Super Kings with 138 wins. Together with Kolkata Knight Riders, these teams have established sustained dominance across IPL history.

Match scoring analysis shows that the average IPL match produces approximately 318 runs, reflecting the aggressive batting nature of modern T20 cricket. While most matches fall within a relatively consistent scoring range, occasional extreme high-scoring and low-scoring contests highlight the dynamic nature of the format.

The analysis also indicates that IPL growth has been accompanied by increased match volume, expanded franchise participation, and broader venue utilization. Furthermore, a small group of elite players account for a significant proportion of Player of the Match awards, emphasizing the importance of individual match-winning performances.

These findings establish a strong foundation for deeper investigations into team strategy, player impact, venue influence, powerplay efficiency, death-over performance, and predictive modeling in subsequent stages of the project.
